```python
import numpy as np
import random
# from queue import Queue


class Queue:
    def __init__(self, maxSize, arr):
        self.size = maxSize
        self.queue = []
        if(len(arr) < maxSize):
            for i in arr:
                self.queue.enqueue(i)
        else:
            print('error: array given is too long')
        
    def enqueue(self, val):
        if(len(self.queue) == self.size):
            self.queue.pop()
        
        self.queue.insert(0, val)
    
    def dequeue(self):
        self.queue.pop()
    
    def getLen(self):
        return len(self.queue)
    
    def full(self):
        return (len(self.queue) == self.size)
    
    def empty(self):
        return (len(self.queue) == 0)
    
    def printQ(self):
        for i in self.queue:
            print(i)
    
    def clearQ(self):
        while(len(self.queue) > 0):
            self.dequeue()
            
    def getItem(self, index):
        return self.queue[index]
    


class Grid:
    rows = 0
    cols = 0
    alive = []
    
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols
        self.randomGrid(rows,cols)
        self.allCells = self.createAll(rows,cols)

        
    def randomGrid(self, rows, cols):
        num_living = random.randrange(1,(rows*cols), 1)
        for i in range(num_living):
            self.alive.append({'row': random.randrange(0,rows,1), 'col': random.randrange(0,cols,1)})
    
            
    def createAll(self,rows,cols):
        all_ = []
        for i in range(rows):
            for j in range(cols):
                all_.append({'row': i, 'col': j})
        return all_
    
    def removeFromLiving(self,row, col):
        self.alive.remove({'row': row, 'col': col})
        
    def addAlive(self, row, col):
        self.alive.append({'row': row, 'col': col})
        
    def checkIfAlive(self, row, col):
        return {'row': row, 'col': col} in self.alive
    
    def getAlive(self):
        print(len(self.alive))
        print(self.alive)
        
    def getNeighbors(self, cell):
        ns = []
        isTop = cell['row'] == 0
        isTop = cell['row'] == 0
        isRight = cell['col'] == self.cols-1
        isBottom = cell['row'] == self.rows-1
        isLeft = cell['col'] == 0
        if(not isTop):
            ns.append({'row': (cell['row'])-1, 'col': (cell['col'])})
        if(not (isTop or isRight)):
            ns.append({'row': (cell['row'])-1, 'col': (cell['col'])+1})
        if(not isRight):
            ns.append({'row': (cell['row']), 'col': (cell['col'])+1})
        if(not (isBottom or isRight)):
            ns.append({'row': (cell['row'])+1, 'col': (cell['col'])+1})
        if(not isBottom):
            ns.append({'row': (cell['row'])+1, 'col': (cell['col'])})
        if(not (isBottom or isLeft)):
            ns.append({'row': (cell['row'])+1, 'col': (cell['col'])-1})
        if(not isLeft):
            ns.append({'row': (cell['row']), 'col': (cell['col'])-1})
        if(not (isTop or isLeft)):
            ns.append({'row': (cell['row'])-1, 'col': (cell['col'])-1})
        
        return ns
        
        


class Game:
    currState = []
    prevStates = Queue(4, [])
    
    def __init__(self, rows, cols, play):
        self.gameGrid = Grid(rows,cols)
        if(play):
            self.fullGame()
#         self.all = self.gameGrid.allCells
#         self.living = self.gameGrid.alive
        
        
        
    def iterate(self):
        for i in self.gameGrid.allCells:
            life = self.checkStatus(i)
            if(self.gameGrid.checkIfAlive(i['row'], i['col'])):
                if(life == 'dead'):
                    self.gameGrid.removeFromLiving(i['row'], i['col'])
            else:
                if(life == 'alive'):
                    self.gameGrid.addAlive(i['row'], i['col'])
                    
        self.currState = []
        self.currState = self.gameGrid.alive.copy()
        self.prevStates.enqueue(self.gameGrid.alive.copy())    
    
    
    def checkStatus(self, cell):
        neighbors = self.gameGrid.getNeighbors(cell)
        count = 0
        for i in neighbors:
            if i in self.gameGrid.alive:
                count += 1
        
        isAlive = self.gameGrid.checkIfAlive(cell['row'], cell['col'])
        
        if(isAlive):
            if(count == 2 or count == 3):
                return 'alive'
            else:
                return 'dead'
        else:
            if(count == 3):
                return 'alive'
            else:
                return 'dead'
            
    
    def sumRows(self, arr):
        count = 0
        for i in arr:
            count += arr['row']
            
        return count
    
    
    def sumCols(self, arr):
        count = 0
        for i in arr:
            count += arr['col']
            
        return count
            
    
    def checkState(self):
        indexToCheck = 0
        if(not self.prevStates.full()):
            return False
        
        
        sum1 = 0
        sum2 = 0
        if(len(self.prevStates.getItem(0)) == len(self.prevStates.getItem(2))):
            sum1 = sumRows(self.prevStates.getItem(0)) + sumCols(self.prevStates.getItem(0))
            sum2 = sumRows(self.prevStates.getItem(2)) + sumCols(self.prevStates.getItem(2))
            indexToCheck = 2
        elif(len(self.prevStates.getItem(0)) == len(self.prevStates.getItem(3))):
            sum1 = sumRows(self.prevStates.getItem(0)) + sumCols(self.prevStates.getItem(0))
            sum2 = sumRows(self.prevStates.getItem(3)) + sumCols(self.prevStates.getItem(3))
            indexToCheck = 3
        
        if((sum1 == sum2) and (sum1 != 0)):
            return fullCheck(indexToCheck)
        
        return False
    
    
    def fullCheck(index):
        for i in self.prevStates.getItem(0):
            if(self.prevStates.getItem(0) != self.prevStates.getItem(index)):
                return False
            
        return True
                
    
    
    def fullGame(self):
        count = 0
        while True:
            self.iterate()
            if(self.checkState()):
                return 'infinite'
            elif(len(self.gameGrid.alive) == 0):
                return 'zero'
            
            print(count)
            count += 1
            
            if(count > 20000):
                return 'infinite'
        
        
    
        

def createUniverse(n, m):
#     test = Grid(n,m)
#     del test
#     test.getAlive()
    testGame = Game(n, m, False)
#     testGame.gameGrid.getAlive()
#     testGame.iterate()
#     testGame.gameGrid.getAlive()
#     testGame.iterate()
#     testGame.gameGrid.getAlive()
    testGame.fullGame()
    
    
#     del testGame
    

createUniverse(29,60)
# createUniverse(29,60)
    
    
```
